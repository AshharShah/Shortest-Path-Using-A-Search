# Initialize The Heuristic Value Table

```python
heuristic_value = {
    'Arad': 366,
    'Zerind': 374,
    'Oradea': 380,
    'Sibiu': 253,
    'Timisoara': 329,
    'Lugoj': 244,
    'Mehadia': 241,
    'Drobeta': 242,
    'Craiova': 160,
    'Rimnicu': 193,
    'Fagaras': 176,
    'Pitesti': 100,
    'Bucharest': 0,
    'Giurgiu': 77,
    'Urziceni': 80,
    'Hirsova': 151,
    'Eforie': 161,
    'Vaslui': 199,
    'Iasi': 226,
    'Neamt': 234
}
```

```python
def sortKey(x):
    return x[0][1]
```

```python
def get_child(current, visited):
    valid_children = []
    for item in graph[current]:
        if(item not in visited):
            valid_children.append(item)
    return valid_children
```

# A-Star Algorithm Implementation

```python
def a_star(start, end):
    # create a priority queue and a visited list
    queue = []
    visited = []
    # create a dictionary to store goal values
    distance = {}
    # add starting node
    queue.append([[start, 0], []])
    # set the g_cost of starting node to 0
    distance[start] = 0
    found = False
    
    while(queue):
        
        print(end  = "\n\n")
        print("Queue Status: " + str(queue))
        # pop the node with lowest f_cost
        current = queue.pop(0)
        # make a copy of the parents till the current node
        route = current[1]
        routes = copy.deepcopy(route)
        # add the current node to the route history for the child nodes
        routes.append(current[0][0])

        # add the current node to the closed list
        visited.append(current[0][0])
        print("Current Node Poped: " + str(current[0][0]))
        
        # get all the neighbouring nodes of the current node
        neighbours = graph[current[0][0]]
        
        # check if the current node is the goal node
        if(current[0][0] == end):
            found = True
            print(end = "\n\n\n")
            print("Goal Node Found!!")
            print("Shortest Path: " + str(routes))
            print("Cost To Reach: " + str(current[0][1]))
            break
        
        # get all the children of the current node that are not visited yet
        childs = get_child(current[0][0], visited)
        
        # for each child node of the current node:
        for child in childs:
            # 1) find the g_cost from dictionary(parent) + the cost to reach child from parent
            g_cost = distance[current[0][0]] + neighbours[child]
            # 2) check if the child is not visited or has a lower g_cost than current one
            if(child not in distance or g_cost < distance[child]):
                # 3) update the g_cost in the dictionary
                distance[child] = g_cost
                # 4) find the f_cost
                f_cost = g_cost + heuristic_value[child]
                # 5) add it to the open list
                queue.append([[child, f_cost], routes])
        # sort the queue based on the f_cost of the nodes
        queue.sort(key = sortKey)
        
        print("Children Of Node: " + str(childs))
```
